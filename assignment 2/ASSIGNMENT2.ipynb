#  Load the Dataset

```python
import pandas as pd

df = pd.read_csv('House Price India.csv')
```

#  Load the Dataset

### Histogram for a numerical variable

```python
import matplotlib.pyplot as plt

plt.hist(df['living area'], bins=20)
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.title('Histogram of Living Area')
plt.show()
```

###  Bar chart for a categorical variable

```python
import seaborn as sns

sns.countplot(x='number of bedrooms', data=df)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Bar Chart of Number of Bedrooms')
plt.show()
```

# Bivariate Analysis

### Scatter plot for two numerical variables

```python
plt.scatter(df['living area'], df['Price'])
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.title('Scatter Plot between Living Area and Price')
plt.show()
```

###  Box plot for comparing a numerical variable across categories

```python
sns.boxplot(x='number of bedrooms', y='Price', data=df)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Box Plot of Price by Number of Bedrooms')
plt.show()
```

# Multivariate Analysis

### Pair plot for multiple numerical variables

```python
sns.pairplot(df[['living area', 'number of bedrooms', 'Price']])
plt.title('Pair Plot of Living Area, Number of Bedrooms, and Price')
plt.show()
```

###  Correlation heatmap for numerical variables

```python
correlation_matrix = df[['living area', 'number of bedrooms', 'Price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Descriptive Statistics

```python
# Calculate descriptive statistics for numerical columns
desc_stats = df.describe()

# Display the descriptive statistics
print(desc_stats)
```

# Handling Missing Values

### Check for Missing Values:

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Display the count of missing values for each column
print(missing_values)
```

### Option 1: Remove Rows with Missing Values:

```python
# Remove rows with missing values
df_cleaned = df.dropna()

# Verify that missing values have been removed
print(df_cleaned.isnull().sum())
```

### Option 2: Fill Missing Values:

```python
# Fill missing values with the mean of the respective column
df_filled = df.fillna(df.mean())

# Verify that missing values have been filled
print(df_filled.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer instance (e.g., filling with mean)
imputer = SimpleImputer(strategy='mean')

# Fit the imputer to your data and transform it
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Verify that missing values have been imputed
print(df_imputed.isnull().sum())
```
